```python
def water_jug_problem(jug1_capacity, jug2_capacity, target):
    steps = []
    jug1 = 0
    jug2 = 0

    while jug1 != target and jug2 != target:
        if jug1 == 0:
            jug1 = jug1_capacity
            steps.append((jug1, jug2))
        elif jug2 == jug2_capacity:
            jug2 = 0
            steps.append((jug1, jug2))
        else:
            pour_amount = min(jug1, jug2_capacity - jug2)
            jug1 -= pour_amount
            jug2 += pour_amount
            steps.append((jug1, jug2))

    # Ensure the other jug has 0 amount of water at the final step
    if jug1 == target:
        steps.append((jug1, 0))
    else:
        steps.append((0, jug2))

    return steps

# Example usage
jug1_capacity = 4
jug2_capacity = 3
target = 2
steps = water_jug_problem(jug1_capacity, jug2_capacity, target)
print("Steps:")
for i, (jug1, jug2) in enumerate(steps):
    print(f"Step {i+1}: Jug1 = {jug1}, Jug2 = {jug2}")
```

```python
def print_board(board):
    for row in board:
        print(" | ".join(row))
        print("-" * 10)
    print()

def check_winner(board, player):
    # Check rows
    for row in board:
        if all(s == player for s in row):
            return True

    # Check columns
    for col in range(3):
        if all(board[row][col] == player for row in range(3)):
            return True

    # Check diagonals
    if all(board[i][i] == player for i in range(3)):
        return True
    if all(board[i][2 - i] == player for i in range(3)):
        return True

    return False

def is_board_full(board):
    return all(all(cell != ' ' for cell in row) for row in board)

def tic_tac_toe():
    board = [[' ' for _ in range(3)] for _ in range(3)]
    players = ['X', 'O']
    current_player = 0

    while True:
        print_board(board)
        try:
            row = int(input(f"Player {players[current_player]}, enter the row (0-2): "))
            col = int(input(f"Player {players[current_player]}, enter the column (0-2): "))
        except ValueError:
            print("Invalid input. Please enter numbers between 0 and 2.")
            continue

        if row < 0 or row > 2 or col < 0 or col > 2:
            print("Invalid move. Indices should be between 0 and 2.")
            continue

        if board[row][col] != ' ':
            print("This cell is already taken. Try again.")
            continue

        board[row][col] = players[current_player]

        if check_winner(board, players[current_player]):
            print_board(board)
            print(f"Player {players[current_player]} wins!")
            break

        if is_board_full(board):
            print_board(board)
            print("The game is a tie!")
            break

        current_player = 1 - current_player

tic_tac_toe()
```

```python
# Find the position of the blank (0)
def find_blank(state):
    for i in range(3):
        for j in range(3):
            if state[i][j] == 0:
                return i, j

# Generate new states by moving the blank (0)
def get_new_states(state):
    x, y = find_blank(state)
    directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]
    new_states = []

    for dx, dy in directions:
        nx, ny = x + dx, y + dy
        if 0 <= nx < 3 and 0 <= ny < 3:
            new_state = [row[:] for row in state]
            new_state[x][y], new_state[nx][ny] = new_state[nx][ny], new_state[x][y]
            new_states.append(new_state)

    return new_states

# BFS algorithm to solve the 8-puzzle problem
def bfs(initial_state):
    queue = [(initial_state, [])]
    visited = set()
    visited.add(tuple(map(tuple, initial_state)))

    while queue:
        state, path = queue.pop(0)
        if state == goal_state:
            return path

        for new_state in get_new_states(state):
            state_tuple = tuple(map(tuple, new_state))
            if state_tuple not in visited:
                visited.add(state_tuple)
                queue.append((new_state, path + [new_state]))

    return None

# Print the state
def print_state(state):
    for row in state:
        print(' '.join(map(str, row)))
    print()

# Example usage

goal_state = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 0]
]

initial_state = [
    [1, 2, 3],
    [4, 0, 6],
    [7, 5, 8]
]

solution_path = bfs(initial_state)

if solution_path:
    print("Solution found!")
    for step, state in enumerate(solution_path):
        print(f"Step {step + 1}:")
        print_state(state)
else:
    print("No solution found.")
```

```python
import math
def minimax (curDepth, nodeIndex, maxTurn, scores, targetDepth):
    # base case : targetDepth reached
    if (curDepth == targetDepth): 
        return scores[nodeIndex]
    
    if (maxTurn):
        return max(minimax(curDepth + 1, nodeIndex * 2, False, scores, targetDepth), 
                   minimax(curDepth + 1, nodeIndex * 2 + 1, False, scores, targetDepth))
    else:
        return min(minimax(curDepth + 1, nodeIndex * 2, True, scores, targetDepth), 
                   minimax(curDepth + 1, nodeIndex * 2 + 1, True, scores, targetDepth))
# Driver code
scores = [3, 5, 2, 9, 12, 5, 23, 23]
treeDepth = math.log(len(scores), 2)
print("The optimal value is : ", end = "")
print(minimax(0, 0, True, scores, treeDepth))
```

```python
def print_board(board):
    for row in board:
        print(" ".join("Q" if col else "." for col in row))
    print()

def is_safe(board, row, col, N):
    # Check this row on left side
    for i in range(col):
        if board[row][i]:
            return False

    # Check upper diagonal on left side
    for i, j in zip(range(row, -1, -1), range(col, -1, -1)):
        if board[i][j]:
            return False

    # Check lower diagonal on left side
    for i, j in zip(range(row, N), range(col, -1, -1)):
        if board[i][j]:
            return False

    return True

def solve_n_queens(board, col, N):
    if col >= N:
        return True

    for i in range(N):
        if is_safe(board, i, col, N):
            board[i][col] = 1
            if solve_n_queens(board, col + 1, N):
                return True
            board[i][col] = 0

    return False

def solve(N):
    board = [[0] * N for _ in range(N)]
    if not solve_n_queens(board, 0, N):
        print("Solution does not exist")
        return False
    print_board(board)
    return True

# Example usage
N = 8
solve(N)
```

```python
class TSP:
    def __init__(self, num_cities, distances):
        self.num_cities = num_cities
        self.distances = distances

    def nearest_neighbor(self):
        unvisited = set(range(1, self.num_cities))
        current_city = 0
        tour = [current_city]
        total_distance = 0

        while unvisited:
            nearest_city = min(unvisited, key=lambda x: self.distances[current_city][x])
            total_distance += self.distances[current_city][nearest_city]
            current_city = nearest_city
            tour.append(current_city)
            unvisited.remove(current_city)

        # Return to starting city
        total_distance += self.distances[current_city][0]
        tour.append(0)

        return tour, total_distance

# Example usage
if __name__ == "__main__":
    num_cities = 5
    distances = [
        [0, 10, 15, 20, 25],
        [10, 0, 35, 25, 30],
        [15, 35, 0, 30, 10],
        [20, 25, 30, 0, 5],
        [25, 30, 10, 5, 0]
    ]

    tsp_solver = TSP(num_cities, distances)
    tour, total_distance = tsp_solver.nearest_neighbor()

    print("Optimal Tour:", tour)
    print("Total Distance:", total_distance)

    print("\nDistance between cities:")
    for i in range(num_cities):
        for j in range(i + 1, num_cities):
            print(f"Distance between city {i} and city {j}: {distances[i][j]}")
```

```python
def interval_scheduling(intervals):
    sorted_intervals = sorted(intervals, key=lambda x: x[1])
    
    result = [sorted_intervals[0]]

    for interval in sorted_intervals[1:]:
        if interval[0] >= result[-1][1]:
            result.append(interval)
    return result

# Example usage
if __name__ == "__main__":
    intervals = [(1, 3), (2, 4), (3, 6), (5, 7), (6, 9)]

    #max_non_overlapping_intervals = interval_scheduling(intervals)
    
    print("Maximum Non-overlapping Intervals:", interval_scheduling(intervals))
```

```python
def airline_scheduling(flights):
    sorted_flights = sorted(flights, key=lambda x: x[0])

    result = [sorted_flights[0]]

    for flight in sorted_flights[1:]:
        if flight[0] >= result[-1][1]:
            result.append(flight)
    
    return result

# Example usage
if __name__ == "__main__":
    flights = [(1, 3), (2, 4), (5, 7), (3, 6), (8, 9)]

    max_flights = airline_scheduling(flights)
    
    print("Maximum Flights Scheduled:", max_flights)
```

```python
class Flight:
    def __init__(self, flight_number, departure_time, arrival_time, maintenance_time):
        self.flight_number = flight_number
        self.departure_time = departure_time
        self.arrival_time = arrival_time
        self.maintenance_time = maintenance_time

def convert_to_minutes(time_str):
    hours, minutes = map(int, time_str.split())
    return hours * 60 + minutes

def schedule_flights(flights):
    flights.sort(key=lambda x: x.arrival_time)
    schedule = []
    current_time = 0
    for flight in flights:
        if flight.departure_time >= current_time + flight.maintenance_time:
            schedule.append(flight)
            current_time = flight.arrival_time
    return schedule

def get_flights_from_user():
    flights = []
    while True:                                              
        flight_number = input("Enter flight number (or 'done' to finish): ")
        if flight_number.lower() == 'done':
            break
        departure_time_str = input("Enter departure time (HH MM): ")
        departure_time = convert_to_minutes(departure_time_str)
        arrival_time_str = input("Enter arrival time (HH MM): ")
        arrival_time = convert_to_minutes(arrival_time_str)
        maintenance_time = int(input("Enter maintenance time (in minutes): "))
        flights.append(Flight(flight_number, departure_time, arrival_time, maintenance_time))
    return flights

if __name__ == "__main__":
    # Get flights from the user
    flights = get_flights_from_user()
    
    # Schedule the flights
    scheduled_flights = schedule_flights(flights)
    
    # Print the result
    print("Scheduled Flights:")
    for flight in scheduled_flights:
        print(f"Flight Number: {flight.flight_number}, Departure Time: {flight.departure_time}, Arrival Time: {flight.arrival_time}")
```
